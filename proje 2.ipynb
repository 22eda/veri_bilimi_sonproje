```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  

R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar. 
State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
Profit (Kâr): Şirketin elde ettiği toplam kâr.  
Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
# Kodu buraya yazınız 

plt.figure(figsize=(10, 6))
plt.scatter(data['R&D Spend'], data['Profit'], alpha=0.6, c='blue', edgecolors='black', s=100)
plt.xlabel('Ar-Ge Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('Ar-Ge Harcaması ve Kâr Arasındaki İlişki', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('gorev1_rd_kar.png', dpi=300, bbox_inches='tight')
plt.show()



```

## 2.GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
# Kodu buraya yazınız 

plt.figure(figsize=(10, 6))
plt.scatter(data['Administration'], data['Profit'], alpha=0.6, c='green', edgecolors='black', s=100)
plt.xlabel('Yönetim Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('Yönetim Harcaması ve Kâr Arasındaki İlişki', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('gorev2_yonetim_kar.png', dpi=300, bbox_inches='tight')
plt.show()




```

## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıran bir çubuk grafik.

```python
# Kodu buraya yazınız 


plt.figure(figsize=(10, 6))
avg_profit_by_state = data.groupby('State')['Profit'].mean().sort_values(ascending=False)
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
bars = plt.bar(avg_profit_by_state.index, avg_profit_by_state.values, 
               color=colors, edgecolor='black', linewidth=1.5)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xlabel('Eyalet', fontsize=12)
plt.ylabel('Ortalama Kâr ($)', fontsize=12)
plt.title('Eyaletlere Göre Ortalama Kâr Karşılaştırması', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('gorev3_eyalet_kar.png', dpi=300, bbox_inches='tight')
plt.show()

```

## 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının dağılımını karşılaştıran bir kutu grafiği.

```python
# Kodu buraya yazınız 


plt.figure(figsize=(12, 6))
spending_data = [data['R&D Spend'], data['Administration'], data['Marketing Spend']]
labels = ['Ar-Ge\nHarcaması', 'Yönetim\nHarcaması', 'Pazarlama\nHarcaması']

bp = plt.boxplot(spending_data, labels=labels, patch_artist=True,
                 boxprops=dict(facecolor='lightblue', alpha=0.7),
                 medianprops=dict(color='red', linewidth=2),
                 whiskerprops=dict(color='black', linewidth=1.5),
                 capprops=dict(color='black', linewidth=1.5))

# Her kutuya farklı renk ver
colors = ['#FF6B6B', '#4ECDC4', '#FFA07A']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

plt.ylabel('Harcama Miktarı ($)', fontsize=12)
plt.title('Harcama Türlerinin Dağılım Karşılaştırması', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('gorev4_harcama_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()



```
